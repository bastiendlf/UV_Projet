```python
from matplotlib import pyplot as plt
import cv2
import os
```

```python
from utils import customJpegCompressor
```

```python
data_folder = '../data/'
pictures = os.listdir(data_folder)

img = cv2.imread(data_folder + pictures[1], 0)

Q_VALUES = ['Q10', 'Q50', 'Q90', 'Q1']
res = []
```

```python
for q in Q_VALUES:
    res.append(customJpegCompressor.compress_decompress(img, q_name=q))
```

```python
plt.figure(figsize=(30, 20))
plt.subplot(2, 3, 1)
plt.imshow(img, cmap='gray')
plt.title("Uncompressed image")


plt.subplot(2, 3, 2)
plt.imshow(res[0], cmap='gray')
plt.title("Compressed jpeg image with Q = q10")

plt.subplot(2, 3, 3)
plt.imshow(res[1], cmap='gray')
plt.title("Compressed jpeg image with Q = q50")

plt.subplot(2, 3, 4)
plt.imshow(res[2], cmap='gray')
plt.title("Compressed jpeg image with Q = q90")


plt.subplot(2, 3, 5)
plt.imshow(res[3], cmap='gray')
plt.title("Compressed jpeg image with Q = q100")

# plt.savefig(f"../output/{pictures[1].split('.')[0]}_qmatrice_comparison.png", facecolor = 'white', dpi = 200)
plt.show()
```

```python
artefacts = [img - resQ for resQ in res]

plt.figure(figsize=(15, 15))
plt.subplot(2, 2, 1)
plt.imshow(artefacts[0], cmap='gray', vmin=-128, vmax=127)
plt.title("Compressed jpeg image with Q = q10")

plt.subplot(2, 2, 2)
plt.imshow(artefacts[1], cmap='gray',  vmin=-128, vmax=127)
plt.title("Compressed jpeg image with Q = q50")

plt.subplot(2, 2, 3)
plt.imshow(artefacts[2], cmap='gray',  vmin=-128, vmax=127)
plt.title("Compressed jpeg image with Q = q90")

plt.subplot(2, 2, 4)
plt.imshow(artefacts[3], cmap='gray',  vmin=-128, vmax=127)
plt.title("Compressed jpeg image with Q = q100")

plt.show()
```
